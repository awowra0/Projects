```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv("COVID-worldometers-data.csv", sep='\t')
#data
```

```python
data_nn = data.dropna(axis=0, ignore_index=True)
#data_nn
```

```python
data_e = data[["Continent","Cases/1M","Country"]]
country_name = 'Europe'
data_e = data_e[data_e['Continent']==country_name]
data_e = data_e.dropna(axis=0, ignore_index=True)
data_e = data_e.sort_values('Cases/1M', ignore_index=True)
fig = px.bar(data_e, x='Cases/1M', y='Country', height=720,
             color=["Less than 100 000" if data_e['Cases/1M'][i]<10**5 else "Between 100 000-500 000" if data_e['Cases/1M'][i]<5*10**5 else "More than 500 000" for i in range(len(data_e))],
             color_discrete_sequence=['steelblue','seagreen', 'tomato'])
fig.update_layout(title=f'Covid - total cases per 1M population in {country_name}', title_x=0.5, title_xanchor='center', title_font=dict(family='Arial', size=20),
                  font_family="Times New Roman", font_size=14, font_color='darkslateblue',
                  legend=dict(title='Count of cases per 1M population', title_font_family='Arial'))
fig.update_xaxes(title_font_family="Arial", title_font_size=18)
fig.update_yaxes(title_font_family="Arial", title_font_size=18, showgrid=True, dtick=1)
fig.add_vline(x=10**5, opacity=0.75, line=dict(color='white'))
fig.add_vline(x=5*10**5, opacity=0.75, line=dict(color='white'))
fig.show()
```

```python
data_nn = data[["Cases", "Deaths", "Continent", "Country"]]
data_nn = data_nn.dropna(axis=0, ignore_index=True)
fig = px.scatter(data_nn,'Cases', 'Deaths', color='Continent', hover_name='Country',
                 color_discrete_sequence=['blue', 'tomato', 'darkgreen', 'purple', 'turquoise', 'gold'],
                 log_x=True, log_y=True)
fig.update_layout(title='Covid - total cases and deaths in the world', title_x=0.5, title_xanchor='center', title_font=dict(family='Arial', size=20),
                  font_family="Times New Roman", font_size=14, font_color='darkslateblue',
                  legend=dict(title_font_family='Arial'))
fig.update_xaxes(title_font_family="Arial", title_font_size=18)
fig.update_yaxes(title_font_family="Arial", title_font_size=18)
```

```python
data_nn = data[["Cases", "Deaths", "Continent", "Country"]]
data_nn = data_nn.dropna(axis=0, ignore_index=True)
fig = px.scatter(data_nn,'Cases', 'Deaths', color='Continent', hover_name='Country',
                 color_discrete_sequence=['blue', 'tomato', 'darkgreen', 'purple', 'turquoise', 'gold'],
                 log_x=True, log_y=True, facet_col_wrap=2, facet_col="Continent")
fig.update_layout(title='Covid - total cases and deaths in the world', title_x=0.5, title_xanchor='center', title_font=dict(family='Arial', size=20),
                  font_family="Times New Roman", font_size=14, font_color='darkslateblue',
                  showlegend=False)
fig.update_xaxes(title_font_family="Arial", title_font_size=18)
fig.update_yaxes(title_font_family="Arial", title_font_size=18)
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
fig.show()
```

```python
data_nn = data[["Tests", "Continent", "Country"]]
data_nn = data_nn.dropna(axis=0, ignore_index=True)
fig = px.box(data_nn, x='Continent', y='Tests', log_y=True, hover_name='Country', points='all', color='Continent',
             color_discrete_sequence=['royalblue', 'tomato', 'darkgreen', 'purple', 'lightseagreen', 'goldenrod'])
fig.update_layout(title='Covid - total tests made in the world', title_x=0.5, title_xanchor='center', title_font=dict(family='Arial', size=20),
                  font_family="Times New Roman", font_size=14, font_color='darkblue', showlegend=False)
fig.update_xaxes(title_font_family="Arial", title_font_size=18)
fig.update_yaxes(title_font_family="Arial", title_font_size=18)
fig.show()
```

```python
data_nn = data[["Cases/1M", "Continent","Country"]]
data_nn = data_nn.dropna(axis=0, ignore_index=True)
fig = px.choropleth(data_nn, locationmode='country names', locations='Country', color='Cases/1M',
                    color_continuous_scale=px.colors.sequential.amp)
fig.update_layout(title='Covid - total cases per 1M population in the world', title_x=0.5, title_xanchor='center', title_font=dict(family='Arial', size=20),
                  font_family="Times New Roman", font_size=14, font_color='darkblue',
                  legend=dict(title_font_family='Arial', title_font_size=16))
fig.show()
```
